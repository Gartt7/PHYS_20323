1. Write a ``Particle`` class that can be used to represent a particle with a mass, a 3-d position, and a 3-d velocity.

2. Write a method that can be used to compute the kinetic energy of the particle

3. Write a method that takes another particle as an argument and finds the distance between the two particles

4. Write a method that given a time interval ``dt`` will update the position of the particle to the new position based on the current position and velocity.

5. Write a ``ChargedParticle`` class that inherits from the ``Particle`` class, but also has an attribute for the charge of the particle.


```python
import numpy as np
class Particle:
    def __init__(self, mass, position, velocity):
        self.mass = mass
        self.position = np.array(position)
        self.velocity = np.array(velocity)

    def kinetic_energy(self):
        vx, vy, vz = self.velocity
        speed_squared = np.sum(self.velocity**2)
        return 0.5 * self.mass * speed_squared
    
    def distance_bw(self, other_particle):
        difference = other_particle.position - self.position
        return np.linalg.norm(difference)

    def updated_position(self, dt):
        self.position = self.position + self.velocity * dt
        



# part 5
class ChargedParticle(Particle):
    def __init__(self, mass, position, velocity, charge):
        super().__init__(mass, position, velocity)
        self.charge = charge
```

```python
class ChargedParticle(Particle):
    def __init__(self, mass, position, velocity, charge):
        super().__init__(mass, position, velocity)
        self.charge = charge

    def find_seperation(self, other_particle):
        seperation = self.distance_bw(other_particle)
        if seperation < 0.25:
            print("Interaction")

            if self.charge * other_particle.charge < 0:
                self.velocity = np.array([0,0,0])
                other_particle.velocity=np.array([0,0,0])

                self.charge = 0
                other_particle.charge = 0
                print("Merge")
            else:
                factor_self = (-1 * (self.charge + other_particle.charge) * (self.mass / other_particle.mass))
                factor_other = (-1 * (self.charge + other_particle.charge) * (other_particle.mass / self.mass))

                self.velocity = self.velocity * factor_self
                other_particle.velocity = other_particle.velocity * factor_other

                print("Repel")
```

```python
p1 = ChargedParticle(mass = 1.0, position = [-5, -5, -5], velocity = [1, 1, 1], charge = 0.5)

p2 = ChargedParticle(mass = 1.0, position = [5, 5, 5], velocity = [-1, -1, -1], charge = -0.5)

for step in range(300):
    if p1.charge !=0 and p2.charge != 0:
        p1.find_seperation(p2)

    p1.updated_position(.05)
    p2.updated_position(.05)

    print(f"Step {step}: P1 pos={p1.position}, vel={p1.velocity}")
    print(f"Step {step}: P2 pos={p2.position}, vel={p2.velocity}")
```
